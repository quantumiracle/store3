```python
import numpy as np
import matplotlib.pyplot as plt

actions=np.load('actions.npy')
ini_pos=np.array([0,0])
pos_set=[ini_pos]
curr_pos=ini_pos
for i in range(len(actions)):
    new_pos=curr_pos+np.array(actions[i])
    pos_set.append(new_pos)
    curr_pos=new_pos
[x,y]=np.transpose(pos_set)
plt.plot(x,y)
```
