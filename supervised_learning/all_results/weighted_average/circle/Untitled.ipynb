```python
# weighted average circle prediction
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(20,3))
results=np.load('circle02.npy', allow_pickle=True)
[predict, label]=np.transpose(results)
predict = predict-0.2# callibrate overall error
shift=4
plt.plot(predict[shift:], '-o', color='b', label='Prediction')
plt.plot(label[:-shift], '-o',color='r', label='Ground Truth')
plt.xlabel('Sample', fontsize=13)
plt.ylabel('Radian', fontsize=13)
plt.legend(loc=1)
plt.grid()
plt.show()
plt.savefig('circle_02.pdf', bbox_inches='tight')
# error
abs_error=np.abs(predict[shift:120]- label[:120-shift])
print(np.argwhere(abs_error>5))  # make sure no alignment error
print('mean, std: ', np.mean(abs_error), np.std(abs_error))
```
