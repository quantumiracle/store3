```python
# weighted average pole prediction
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
results=np.load('pole00.npy', allow_pickle=True)
predict=results[:, 0]
label=results[:, 1]
# norm factor for sim-to-real transfer
a0=300
b0=0.001
a1=200
b1=-0.001
shift=1  # frame shift
predict = np.transpose([predict[:, 0]/a0-b0, predict[:, 1]/a1-b1])
# predict = predict/a0-b0
plt.subplot(2,1,1)
plt.plot(predict[shift:,0], '-o', color='b', label='Prediction')
plt.plot(label[:-shift,0], '-o',color='r', label='Ground Truth')
plt.xlabel('Sample', fontsize=13)
plt.ylabel('X-position', fontsize=13)
plt.legend(loc=1)
plt.grid()

plt.subplot(2,1,2)
plt.plot(predict[shift:,1], '-o', color='b', label='Prediction')
plt.plot(label[:-shift,1], '-o',color='r', label='Ground Truth')
plt.xlabel('Sample', fontsize=13)
plt.ylabel('Y-position', fontsize=13)
plt.legend(loc=1)
plt.grid()

plt.savefig('pole_00.pdf', bbox_inches='tight')
plt.show()

# error
abs_error=np.abs(predict[shift:]- label[:-shift])
print(np.argwhere(abs_error>0.1))  # make sure no alignment error
print('mean, std: ', np.mean(abs_error), np.std(abs_error))
```
