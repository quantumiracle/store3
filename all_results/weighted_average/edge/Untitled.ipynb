```python
# weighted average edge prediction
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(20,3))
results=np.load('edge00.npy', allow_pickle=True)
[predict, label]=np.transpose(results)
# norm factor for sim-to-real transfer
a=420
b=-0.002
shift=4  # frame shift
predict = predict/a-b
plt.plot(predict[shift:], '-o', color='b', label='Prediction')
plt.plot(label[:-shift], '-o',color='r', label='Ground Truth')
plt.xlabel('Sample', fontsize=13)
plt.ylabel('X-position', fontsize=13)
plt.legend()
plt.grid()
plt.savefig('edge_00.pdf', bbox_inches='tight')

# error
abs_error=np.abs(predict[shift:]- label[:-shift])
print(np.argwhere(abs_error>0.1))  # make sure no alignment error
print('mean, std: ', np.mean(abs_error), np.std(abs_error))
```
